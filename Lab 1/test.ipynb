## DATA ANALYSIS

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Health_heart_experimental.csv")
df.head(9)
```

## pie chart


```python
df['indication'].value_counts().plot(kind='pie',autopct='%0.2f')
```

#### it shows that 52.68%  peoples are indicated healthy and in safe zone but 47.32% peoples are not in safe zone they may have some health issue or it says they are not healthy

```python
df['sex'].value_counts().plot(kind='pie',autopct='%0.2f')
```

#### similarly this pie chart shows the ratio of male and female in the given data we can see that number of males are female are nearly equal 

## Box Plot

```python
sns.boxplot(df['HR'])
```

#### From the above plot it is clearly viewed that average Heart Rate of person is 140 and first 25% people have nearly 110 Heart Rate which is considerd as healthy and safe and last 25% people have Heart Rate greater than 170.

## Histogram


```python
plt.hist(df['weightKg'])
```

#### The above plot shows the right skewed distribution where it clarifies that the most peoples are 50 kg ie more than 12000 peoples and it gradually decreases to right and the people who are nearly 120 kg are very few nearlt 3000.

```python
plt.hist(df['heightCm'])
```

#### Similarly this histogram plot shows the skewed left distribution of peoples height that means most of the people are tall 180 cm and least of people, nearly 800 are short (110 cm) .
