## Linear Algebra

```python
import numpy as np
```

```python
A = np.array([
    [2,-1],
    [0,3]
])

B = np.array([
    [1,2],
    [-3,4]
])
```

### Addition

```python
print(f" Matrix A + B is\n{A+B}")
```

### Multiplication

```python
# dot product
print("Dot product using 'np.dot' --->\n",np.dot(A,B)) 

print("Dot product using '.dot' --->\n",A.dot(B))

print("Dot product using '@' --->\n",A @ B)
print((A @ B).shape) 
```

```python
print("3A - 2B =\n", 3*A - 2*B)
```

## Determinant and Inverse

```python
import numpy.linalg as LA

determinant = LA.det(A)
print(" |A| = ",determinant)
```

#### Invertible

```python
if determinant != 0:
    print(LA.inv(A))
else:
    print("Matrix is not invertible")
```

## Eigenvalues and Eigenvectors  

```python
print("A =\n", A)
eigenvalues , eigenvectors = np.linalg.eig(A)
print("\n Eigen value 1 is:",eigenvalues[0])
print(" and Eigen vector corresponding is:",eigenvectors[:, 0])
print("\n Eigen value 2 is:",eigenvalues[1])
print(" and Eigen vector corresponding is:",eigenvectors[:, 1])
```

### Prediction


```python
X = np.array([[4,1.5,100],
              [6,2.0,150]])
y = np.array([[20],
              [30]])
beta = LA.pinv(X) @ y # here normal formula [β=(X.T * X)^−1  X.T y] doesnot work because 
                      #in this case (X.T * X)^−1 doesnot exist as the determinant is 0 so pseudo inverse gives the best possible solution
                    
X3 = np.array([[8,3,200]])
price3 = X3 @ beta
print(price3)
```

### In the above example we find the price of a third car with its given feature by looking the trend of data.
